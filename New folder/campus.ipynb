```python
def subject_data(value):
    res="N"
    if value=="A":
        res=1
    elif value=="B":
        res=2
    elif value=="C":
        res=3
    elif value=="S":
        res=4
    else:
        res=5
        
    return res
```

```python
data['O/L Maths'] = data['O/L Maths '].apply(subject_data)
data = data.drop(columns=['O/L Maths '])
data['O/L Science'] = data['O/L Science'].apply(subject_data)
data['O/L English'] = data['O/L  English '].apply(subject_data)
data = data.drop(columns=['O/L  English '])
data['A/L Combine Maths'] = data['A/L Combine Maths'].apply(subject_data)
data['A/L Bio'] = data['A/L Bio'].apply(subject_data)
data['A/L Chemistry'] = data['A/L Chemistry'].apply(subject_data)
data['A/L Physics'] = data['A/L Physics'].apply(subject_data)
```

```python
data['Public/Private'] = data['Public/Private'].apply(np.int64)
data['University'] = data['University'].apply(np.int64)
```

```python
X = data.drop('University', axis=1)
y = data['University']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
X_train.shape, X.shape
```

```python
def model_acc(model):
    model.fit(X_train, y_train)
    acc = model.score(X_test, y_test)
    print(str(model)+ ' --> ' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
```

```python
import pickle
# save model to file
pickle.dump(dt, open("campus.dat", "wb"))
```

```python
predict=np.round(dt.predict(X_test))
```

```python
A = np.array(y_test).reshape(-1, 1)
B = predict.reshape(-1, 1)
print(type(predict))
print(type(y_test))
plt.rcParams['figure.figsize'] = 16,5
plt.figure()
plt.plot(A[-100:], label="Real")
plt.plot(B[-100:], label="Predicted")
plt.legend()
plt.title("Result : real vs predicted")
plt.ylabel("Result")
plt.xticks(())
plt.show()
```
