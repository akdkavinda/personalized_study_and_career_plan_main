```python
data['District']=data['District'].str.replace('Mathara','Matara')
```

```python
def district_data(value):
    res=""
    if value=="Colombo":
        res=1
    elif value=="Rathnapura":
        res=2
    elif value=="Kandy":
        res=3
    elif value=="Galle":
        res=4
    elif value=="Gampaha":
        res=5
    elif value=="Matara":
        res=6
    elif value=="Kalutara":
        res=7
    elif value=="Hambantota":
        res=8
    elif value=="Monaragala":
        res=9
    elif value=="Badulla":
        res=10
    elif value=="Kagalle":
        res=11
    elif value=="Anuradhapura":
        res=12
    elif value=="Mullaitivu":
        res=13
    elif value=="Vavuniya":
        res=14
    elif value=="Trincomalee":
        res=15
    elif value=="Puttalam":
        res=16
    elif value=="Polonnaruwa":
        res=17
    elif value=="Nuwara Eliya":
        res=18
    elif value=="Mannar":
        res=19
    elif value=="Kilinochchi":
        res=20
    elif value=="Jaffna":
        res=21
    elif value=="Batticaloa":
        res=22
    elif value=="Matale":
        res=23
    elif value=="Mannar":
        res=24
    elif value=="Ampara":
        res=25
    elif value=="Kurunegala":
        res=26
        
    return res
```

```python
data['District'] = data['District'].apply(district_data)
```

```python
def university_data(value):
    res=""
    if value=="SLIIT":
        res=1
    elif value=="Ruhuna":
        res=2
    elif value=="Open University":
        res=3
    elif value=="Moratuwa":
        res=4
        
    return res
```

```python
data['University name'] = data['University name'].apply(university_data)
```
