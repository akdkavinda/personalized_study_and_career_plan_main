```python
def CGPA_data(value):
    if value>9.0:
        res=1
    elif value>8.5:
        res=2
    elif value>8.0:
        res=3
    elif value>7.5:
        res=4
    elif value>7.0:
        res=5
    elif value>6.5:
        res=6
    elif value>6.0:
        res=7
    elif value>5.5:
        res=8
    elif value>5.0:
        res=9
    elif value>4.5:
        res=10
    elif value>4.0:
        res=11
    elif value>3.5:
        res=12
    elif value>3.0:
        res=13
    elif value>2.5:
        res=14
    elif value>2.0:
        res=15
    elif value>1.5:
        res=16
    elif value>1.0:
        res=17
    else:
        res=18
        
    return res
```

```python
data['CGPA'] = data['CGPA'].apply(CGPA_data)
```

```python
def yes_no_data(value):
    if value=="yes" or value=="Yes" or value=="YES":
        res=1
    else:
        res=0
        
    return res
```

```python
data['Q1'] = data['Did you do webdev during college time ?'].apply(yes_no_data)
data['Q2'] = data['Are you good at Data analysis ?'].apply(yes_no_data)
data['Q3'] = data['reading and writing skills'].apply(yes_no_data)
data['Q4'] = data['Are you a tech person ?'].apply(yes_no_data)
data['Q5'] = data['Were you in a non tech society ?'].apply(yes_no_data)
data['Q6'] = data['Are you good at coding ?'].apply(yes_no_data)
data['Q7'] = data['Have you developed mobile apps ?'].apply(yes_no_data)
data['Q8'] = data['Are you good at communication ?'].apply(yes_no_data)
data['Q9'] = data['Do you have specialization in security'].apply(yes_no_data)
data['Q10'] = data['Have you ever handled large databases ?'].apply(yes_no_data)
data['Q11'] = data['Do you have knowlege of statistics and data science?'].apply(yes_no_data)
data['Q12'] = data['Are you proficient in English ?'].apply(yes_no_data)
data['Q13'] = data['Have you ever managed some event?'].apply(yes_no_data)
data['Q14'] = data['Do you write technical blogs ?'].apply(yes_no_data)
data['Q15'] = data['Are you into marketing ?'].apply(yes_no_data)
data['Q16'] = data['Are you a ML expert ?'].apply(yes_no_data)
data['Q17'] = data['Do you have a lot of connections ?'].apply(yes_no_data)
data['Q18'] = data['Have you ever built live project ?'].apply(yes_no_data)
data=data.drop({'Did you do webdev during college time ?','Are you good at Data analysis ?'
                ,'reading and writing skills','Are you a tech person ?','Were you in a non tech society ?','Are you good at coding ?',
                'Have you developed mobile apps ?','Are you good at communication ?','Do you have specialization in security',
                'Have you ever handled large databases ?','Do you have knowlege of statistics and data science?','Are you proficient in English ?',
                'Have you ever managed some event?','Do you write technical blogs ?','Are you into marketing ?','Are you a ML expert ?',
                'Do you have a lot of connections ?','Have you ever built live project ?'}, axis=1)
```

```python
def Role_data(value):
    if value=="Security":
        res=10
    elif value=="Content Writer":
        res=2
    elif value=="Marketing":
        res=3
    elif value=="ML engineer":
        res=4
    elif value=="Management":
        res=5
    elif value=="Computer Analyst":
        res=6
    elif value=="Data Analysis":
        res=7
    elif value=="Network Engineer":
        res=8
    elif value=="Data Engineer":
        res=9
    else:
        res=1
        
    return res
```

```python
data['Result'] = data['Role'].apply(Role_data)
```

```python
data=data.drop({'Role'}, axis=1)
```

```python
data
```

```python
X = data.drop('Result', axis=1)
y = data['Result']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
X_train.shape, X.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
```

```python
reg1 = lr.fit(X_train, y_train)

reg1.score(X_test, y_test)
```

```python
reg2 = rf.fit(X_train, y_train)

reg2.score(X_test, y_test)
```

```python
import pickle

pickle.dump(reg1, open("role_model.dat", "wb"))
```
